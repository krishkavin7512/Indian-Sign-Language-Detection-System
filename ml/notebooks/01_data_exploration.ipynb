# ISL Dataset Exploration

Explore all four datasets: Kaggle ISL, INCLUDE, ISL-CSLTR, iSign.

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from pathlib import Path
import pandas as pd

DATASETS = Path('../../datasets')
print('Dataset root:', DATASETS.resolve())
```

```python
# --- Kaggle ISL static dataset ---
kaggle_path = DATASETS / 'kaggle_isl'
classes = sorted([d.name for d in kaggle_path.iterdir() if d.is_dir()]) if kaggle_path.exists() else []
print(f'Kaggle ISL classes ({len(classes)}): {classes}')
for cls in classes[:5]:
    imgs = list((kaggle_path / cls).glob('*.jpg')) + list((kaggle_path / cls).glob('*.png'))
    print(f'  {cls}: {len(imgs)} images')
```

```python
# --- INCLUDE dataset ---
include_path = DATASETS / 'include'
if include_path.exists():
    folders = sorted([d for d in include_path.iterdir() if d.is_dir()])
    video_counts = {f.name: len(list(f.glob('*.mp4'))) for f in folders}
    print(f'INCLUDE classes: {len(folders)}')
    print(f'Total videos: {sum(video_counts.values())}')
    df = pd.DataFrame(list(video_counts.items()), columns=['Class', 'Videos'])
    print(df.describe())
```

```python
# Sample frames from INCLUDE dataset
include_path = DATASETS / 'include'
if include_path.exists():
    sample_classes = sorted([d for d in include_path.iterdir() if d.is_dir()])[:9]
    fig, axes = plt.subplots(3, 3, figsize=(12, 9))
    for ax, cls_dir in zip(axes.flatten(), sample_classes):
        videos = list(cls_dir.glob('*.mp4'))
        if videos:
            cap = cv2.VideoCapture(str(videos[0]))
            ret, frame = cap.read()
            cap.release()
            if ret:
                ax.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
                ax.set_title(cls_dir.name, fontsize=10)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```
