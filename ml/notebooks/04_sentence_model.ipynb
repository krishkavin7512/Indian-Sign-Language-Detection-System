# Sentence Model — Conv1D + BiLSTM + CTC

Train on ISL-CSLTR: 700 continuous sentence videos for sequence-to-sequence ISL recognition.

```python
import numpy as np
import json
from pathlib import Path

DATA_PATH = Path('../data/processed')
try:
    seqs = np.load(DATA_PATH / 'sentence_sequences.npy', allow_pickle=True)
    with open(DATA_PATH / 'sentence_labels.json') as f:
        labels = json.load(f)
    with open(DATA_PATH / 'sentence_vocab.json') as f:
        vocab = json.load(f)
    print(f'Sequences: {len(seqs)}')
    print(f'Vocabulary size: {len(vocab)}')
    print(f'Sample: "{labels[0]}"  → seq shape: {seqs[0].shape}')
except FileNotFoundError:
    print('Run preprocess_sentence.py first')
```
