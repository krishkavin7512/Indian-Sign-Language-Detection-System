# Static Sign Classifier — Training & Evaluation

Train MLP on 63-dim MediaPipe hand landmarks for A-Z and 0-9 recognition.

```python
import numpy as np
import json
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path('../data/processed')
if not DATA_PATH.exists():
    print('Run preprocess_static.py first')
else:
    X_train = np.load(DATA_PATH / 'static_X_train.npy')
    y_train = np.load(DATA_PATH / 'static_y_train.npy')
    print(f'Train set: {X_train.shape}, {len(np.unique(y_train))} classes')
```

```python
# Visualize class distribution
with open(DATA_PATH / 'static_label_map.json') as f:
    label_map = json.load(f)
import seaborn as sns
label_names = [label_map[str(i)] for i in y_train]
uniq, counts = zip(*sorted(zip(*np.unique(label_names, return_counts=True)), key=lambda x: -x[1]))
fig, ax = plt.subplots(figsize=(16, 4))
ax.bar(uniq, counts, color='steelblue')
ax.set_title('Class Distribution')
plt.tight_layout()
plt.show()
```
